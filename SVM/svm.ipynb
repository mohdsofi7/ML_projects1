```python
X = data.drop('Outcome',axis=1)
```

```python
y =data['Outcome']
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=123,stratify=y)
```

```python
X_train
```

```python
print(pd.value_counts(y_train,normalize=True)*100)
print(pd.value_counts(y_test,normalize=True)*100)
```

```python
from sklearn.svm import SVC
```

```python
svm_model = SVC(kernel='rbf', C=1.0)
```

```python
svm_model.fit(X_train,y_train)
```

```python
y_pred = svm_model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score,classification_report
```

```python
print("Confusion Matrix \n", confusion_matrix(y_test,y_pred))
print("Accurcay : ", accuracy_score(y_test, y_pred))
print("Recall   : ", recall_score(y_test, y_pred))
print("Precision: ", precision_score(y_test, y_pred)) 
```

## Hyper parameter tuning for SVC

```python
from sklearn.model_selection import GridSearchCV 

param_grid = {'C': [0.1, 1, 10, 100, 1000], 
			'gamma': [1, 0.1, 0.01, 0.001, 0.0001], 
			'kernel': ['rbf']} 

grid = GridSearchCV(svm_model, param_grid, refit = True, verbose = 3) 
 
grid.fit(X_train, y_train)
```

```python
print(grid.best_params_) 
```

```python
grid_predictions = grid.predict(X_test) 
```

```python
print(classification_report(y_test, grid_predictions))
```

```python
print("Confusion Matrix \n", confusion_matrix(y_test,grid_predictions))
print("Accurcay : ", accuracy_score(y_test, grid_predictions))
print("Recall   : ", recall_score(y_test, grid_predictions))
print("Precision: ", precision_score(y_test, grid_predictions)) 
```
